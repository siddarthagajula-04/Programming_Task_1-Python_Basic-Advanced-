## Running Sum

```python
from typing import List
```

```python
class RunningSum:
    def runningSum(self, nums: List[int]) -> List[int]:
        total = 0
        op = []
        for i in nums:
            total += i
            op.append(total)
        return op
```

```python
RS = RunningSum()
RS.runningSum([3,1,2,10,1])
```

## Shuffle the Array

```python
class Shuffle:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        op = []
        for i in range(n):
            op.append(nums[i])
            op.append(nums[i+n])
        return op
```

```python
sf = Shuffle()
sf.shuffle([2,5,1,3,4,7], 3)
```

## Kids with the greatest number of candies

```python
class KidsWithCandies:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        res = []
        max_val = max(candies)
        for i in candies:
            val = i+extraCandies
            if max_val > val:
                res.append(False)
            else:
                res.append(True)
        return res
```

```python
KWC = KidsWithCandies()
KWC.kidsWithCandies([2,3,5,1,3], 3)
```
