## Python If-Else

```python
n = int(input())

if n % 2 != 0:     # odd
    print("Weird")
elif n % 2 == 0 and 2 <= n <= 5:   #even
    print("Not Weird")
elif n % 2 == 0 and 6 <= n <= 20:    #even
    print("Weird")
elif n % 2 == 0 and n > 20:    #even
    print("Not Weird")
```

## Write a Function

```python
def is_leap(year):
    leap = False
    
    if (year % 400 == 0):
        leap = True
    elif (year % 100 == 0):
        leap = False
    elif (year % 4 == 0):
        leap = True
    
    return leap

year = int(input())
print(is_leap(year))
```
